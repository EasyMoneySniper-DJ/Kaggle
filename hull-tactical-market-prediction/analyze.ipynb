```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("train.csv")
data["lagged_forward_returns"] = data["forward_returns"].shift(1)
data["lagged_risk_free_rate"] = data["risk_free_rate"].shift(1)
data["lagged_market_forward_excess_returns"] = data["market_forward_excess_returns"].shift(1)
data
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data, x="E3", kde=True, color="skyblue")
plt.show()
```

```python
data.drop(["date_id"], axis=1).hist(figsize=(30,30), bins=30, edgecolor='black')
plt.tight_layout()
plt.show()
```
